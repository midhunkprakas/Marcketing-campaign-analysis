### Outlier Analysis

Outliers were observed in key variables such as income and customer spending. 
These extreme values represent high-value customers and were carefully evaluated before deciding whether to retain or remove them to ensure meaningful analysis.

```python
## Handling Outliers

Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)

lower = Q1 - 1.5*(Q3 - Q1)
upper = Q3 + 1.5*(Q3 - Q1)

df['Income'] = df['Income'].clip(lower, upper)

```

```python
outliers = df[(df['Income'] < lower) | (df['Income'] > upper)]
```

## <span style="color:#4CAF50;"> 🔗 Bivariate Analysis (Relationships) </span>

```python
plt.hexbin(df['Income'], df['Total_Spending'], gridsize=25)
plt.xlabel("Income")
plt.ylabel("Total Spending")
plt.title("Income vs Spending Density")
plt.show()
```

**Insight**

```python
sns.boxplot(x=pd.cut(df['Age'], bins=5), y=df['Total_Spending'])
plt.title("Age Group vs Spending")
plt.show()
```



**Insights**

There is a clear **positive correlation** where total spending increases progressively with age. Customers in the oldest age bracket (approx. 75–85) exhibit the highest median spending and most consistent behavior, while younger groups show lower median spending but higher variability with several high-value outliers.

**Business Interpretations**

Prioritize the **senior segment** as the primary revenue driver, focusing on premium loyalty programs and high-ticket items. For the younger demographic, implement **predictive modeling** to identify the "high-potential" outliers and nurture them into long-term high-value customers through personalized engagement.

## <span style="color:#4CAF50;"> 👨‍👩‍👧 Family Impact </span>

## <span style="color:#4CAF50;"> 🛒 Channel Analysis</span>

```python
channels = df[['NumWebPurchases','NumStorePurchases','NumCatalogPurchases']].sum()
channels.plot(kind='pie', autopct='%1.1f%%')
plt.title("Purchase Channel Share")
plt.ylabel('')
plt.show()
```

### Analysis Breakdown: Purchase Channel Share

**Insights**

Physical stores are the dominant sales channel (46.2%), followed by web purchases (32.5%), while catalogs maintain a significant niche share of 21.3%. This indicates a customer base that still heavily values the tactile experience of in-person shopping alongside digital convenience.

**Business Interpretations**

Prioritize **omnichannel strategies** that use stores as high-touch experience centers while optimizing the website for seamless transactions. Catalog marketing should not be abandoned but rather repurposed as a **high-end, targeted outreach tool** for the specific 21% segment that prefers traditional browsing.

## <span style="color:#4CAF50;"> 🔥 Correlation Analysis</span>

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True),cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```



### Analysis Breakdown: Correlation Heatmap

**Insights**

  * **Primary Spend Drivers:** Income shows the strongest positive correlation with total spending, particularly in the **Wine (MntWines)** and **Meat (MntMeatProducts)** categories.
  * **The Family Constraint:** A significant negative correlation exists between the **number of children (Kidhome)** and total spending, indicating that household size is a major barrier to high-value purchases in this dataset.
  * **Purchasing Synergy:** High correlations between **Catalog, Web, and Store purchases** suggest that high-spenders are "omnichannel"—they don't stick to one platform but buy across all available touchpoints.

**Business Interpretations & Solutions**

  * **Upsell Strategy:** Since wine and meat are "gateway" high-value categories, bundle these items together or use them as lead magnets for your high-income segments.
  * **Channel Optimization:** Since web visits (NumWebVisitsMonth) show a weak or negative correlation with actual conversion, **redesign the digital UI/UX** to reduce friction and mimic the higher success rates of store and catalog channels.
  * **Segmented Relief:** To capture the "low-spending" family segment, introduce **family-tier loyalty rewards** or bulk-purchase discounts specifically for staples to offset the negative correlation between children and discretionary spend.

## <span style="color:#4CAF50;"> 📢 Campaign Analysis</span>

```python
fig = px.pie(df, names='Response', title="Campaign Response")
fig.show()
```

### Analysis Breakdown: Campaign Response

**Insights**

  * **Low Conversion Rate:** The campaign response  chart shows a **15% acceptance rate**, indicating that while a core group is engaged, the vast majority (85%) of the current audience remains unresponsive to standard marketing prompts.
  * **Segmented Success:** Combined with your segmentation scatter plot, it is clear that the 15% who responded are primarily **high-income, high-spending individuals**, suggesting the campaign resonated with premium buyers rather than the general mass market.

**Business Interpretations & Solutions**

  * **Targeting Refinement:** Stop "blanket" marketing to the 85% non-responsive group to save costs; instead, use **predictive lookalike modeling** to find more customers who match the profile of the 15% successful converters.
  * **Offer Personalization:** Since the current campaign failed for the majority, test **lower-barrier entry offers** (like discounts on smaller quantities) for the mid-to-low income segments who were likely alienated by high-ticket campaign items.

```python
fig = px.scatter(df, x="Income", y="Total_Spending",
                 size="Total_Spending",
                 color="Response",
                 hover_data=['Age'],
                 title="Customer Segmentation")
fig.show()
```

### Analysis Breakdown: Customer Segmentation

**Insights**
* **The "Profitability Sweet Spot":** High-spending customers are concentrated in the **\$60k – \$100k income bracket**, showing a dense cluster of individuals who spend over \$1,000.
* **Response Concentration:** The visual clearly shows that **campaign responders (yellow dots)** are almost exclusively located in the high-income/high-spending quadrant, while low-spenders (dark dots) rarely engage.
* **Outlier Opportunity:** A few low-income but high-spending "aspirational" outliers exist, suggesting that for some customers, spending is driven by lifestyle preference rather than just pure disposable income.

**Business Interpretations & Solutions**
* **Premium Tier focus:** Formalize a **"Platinum Segment"** for those in the top-right quadrant (\$80k+ income) with exclusive early access to products, as they have the highest probability of accepting offers.
* **Value Proposition Shift:** Since low-income groups show very low response rates, the current marketing "language" is likely too premium; create a **secondary brand track** that emphasizes utility and cost-savings to capture the lower-left quadrant.
* **Behavioral Targeting:** Don't just target by income; use **Spending-to-Income ratios** to identify "enthusiast" customers who spend a high percentage of their earnings, as they are your most loyal and responsive advocates regardless of their exact pay bracket.
